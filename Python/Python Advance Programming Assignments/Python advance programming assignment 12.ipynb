Examples

#2+3 = 5, 6+9 = 15
#26 + 39 = 515

meme_sum(122, 81) ➞ 1103
#1+0 = 1, 2+8 = 10, 2+1 = 3
#122 + 81 = 1103

meme_sum(1222, 30277) ➞ 31499



```python
def meme_sum(a,b):
    m=len(str(a))-len(str(b))
    if len(str(a))>len(str(b)):
        f1,f2=str(a),abs(m)*"0"+str(b)
    elif len(str(a))<len(str(b)):
        f1,f2=abs(m)*"0"+str(a),str(b)
    else:
        f1,f2=str(a),str(b)
    temp=0
    l=[]
    for i in range(len(f1)):
        temp=int(f1[i])+int(f2[i])
        l.append(str(temp))
    print("".join(l)) 
meme_sum(26, 39)  
meme_sum(122, 81)
meme_sum(1222, 30277)
```
